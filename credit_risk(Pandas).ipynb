```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("credit_risk_dirty.csv")
```

```python
# Remove Completely Blank Column
df.drop(columns=["remarks"], inplace=True)
```

```python
# Remove Duplicate Rows
df.drop_duplicates(inplace=True)
```

```python
# Remove Leading & Trailing Spaces
df["customer_id"] = df["customer_id"].str.strip()
```

```python
# Standardize Gender Labels
df["gender"] = (
    df["gender"]
    .str.strip()
    .str.lower()
    .replace({
        "m": "Male",
        "male": "Male",
        "man": "Male",
        "f": "Female",
        "female": "Female",
        "woman": "Female"
    })
)
```

```python
# Standardize Loan Purpose
df["loan_purpose"] = (
    df["loan_purpose"]
    .str.strip()
    .str.lower()
    .replace({
        "home loan": "Home Loan",
        "home": "Home Loan",
        "education": "Education",
        "edu": "Education",
        "business": "Business",
        "personal": "Personal"
    })
)
```

```python
# Standardize Default Labels
df["default"] = (
    df["default"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "yes": 1,
        "y": 1,
        "true": 1,
        "1": 1,
        "no": 0,
        "n": 0,
        "false": 0,
        "0": 0
    })
)
```

```python
# Remove Currency Symbols
df["loan_amount"] = (
    df["loan_amount"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
# Convert Numeric Columns
numeric_cols = ["age", "annual_income", "credit_score","loan_tenure_months"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Handle Impossible Values
df["age"] = df["age"].clip(18, 100)
df["annual_income"] = df["annual_income"].clip(lower=0, upper=1000000)
df["credit_score"] = df["credit_score"].clip(300, 850)
df["loan_tenure_months"] = df["loan_tenure_months"].abs()
```

```python
# Fill Missing Values
df["age"] = df["age"].fillna(df["age"].median())
df["annual_income"] = df["annual_income"].fillna(df["annual_income"].median())
df["credit_score"] = df["credit_score"].fillna(df["credit_score"].median())
df["customer_id"] = df["customer_id"].fillna(df["customer_id"].mode()[0])
df["loan_purpose"] = df["loan_purpose"].fillna(df["loan_purpose"].mode()[0])
```

```python
# Convert to numeric
df["employment_years"] = pd.to_numeric(df["employment_years"], errors="coerce")
```

```python
# Keep values between 0 and 50 years
df["employment_years"] = df["employment_years"].clip(lower=0, upper=50)
```

```python
# Fill missing values
df["employment_years"] = df["employment_years"].fillna(df["employment_years"].median())
```

```python
# Convert to integer
df["employment_years"] = df["employment_years"].astype(int)
```

```python
# Save Clean Dataset
df.to_csv("credit_risk_clean.csv", index=False)
print("✅ Data cleaning completed successfully.")
```
